#Apply linear regression on the iris dataset but using Sepal Length as the target and rest of the features as the dataset

```python
import pandas as pd
```

```python
#upload
data=pd.read_csv("Iris.csv")
print(data)
```

```python
#drop the Species   column
data.drop('Species',axis=1,inplace=True)
print(data)
```

```python
#extract SepalLengthCm   as target
target=data.iloc[:,1].values
print(target)
```

```python
data.drop('SepalLengthCm',axis=1,inplace=True)
print(data)
```

```python
#remains extract as features
features=data.iloc[:,:].values
print(features)
```

```python
#split the data set n% for training and remaing for testing
#test size hw much % take fore test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=1)
```

```python
#import a package create an obj call a package for Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model=model.fit(x_train,y_train)
#after train evaluate that model for that we use the function score
model.score(x_train,y_train)
```

```python
#actual perfrmance evaluate by testing data so we do
model.score(x_test,y_test)
```

```python
#compare predicted value with actual
y_train_pred=model.predict(x_train)
```

```python
for x,y in zip(y_train,y_train_pred):
  print(x,y)
```

```python
#done on traing data
from matplotlib import pyplot as plt
plt.plot(y_train)
plt.plot(y_train_pred)
plt.legend(['actual','predicted'])
plt.show()
```

```python
#now for test
y_test_pred=model.predict(x_test)
```

```python
#now vishvalization on testing data 
from matplotlib import pyplot as plt
plt.plot(y_test)
plt.plot(y_test_pred)
plt.legend(['actual','predicted'])
plt.show()
```

```python
import numpy as np
```

```python
#calculate the value of beta
x = x_test
x_T = x.transpose()
x_T_x = x_T.dot(x)
x_T_x_inv = np.linalg.inv(x_T_x)
beta = x_T_x_inv.dot(x_T).dot(y_test)
print(beta)
```

```python
beta_x = beta.dot(x_T)
print(beta_x)
```

```python
plt.plot(np.arange(len(beta_x)),beta_x)
plt.plot(np.arange(len(x_test)),y_test_pred)
plt.plot(np.arange(len(x_test)),y_test)
plt.legend(['Y Calculated', 'Y Predicted','Actual Y'])
plt.title('Y_pred and Y_test')
```
